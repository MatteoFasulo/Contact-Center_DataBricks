## Creazione dei DataFrame e Temporary Table (Punti 4 e 5)

### DataFrame Casi

```
# File location and type
file_location = "/FileStore/LUMSA/Casi.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ";"

# The applied options are for CSV files. For other file types, these will be ignored.
casi = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(casi)

temp_table_name = "Casi_csv"

casi.createOrReplaceTempView(temp_table_name)

# Numero di osservazioni
print(f'Il DF casi contiene: {casi.count()} occorrenze')

# Poiché sia KEY_DATA che DataArrivo sono data non in formato ISO, provvedo a renderle tali per semplificare qualunque futura operazione inerente alle date
# Prima trasformo la struttura a datafram di Pandas attraverso toPandas()
CASI = casi.toPandas()
CASI['KEY_DATA'] = pd.to_datetime(CASI['KEY_DATA'],format='%Y|%m|%d')
CASI['DataArrivo'] = pd.to_datetime(CASI['DataArrivo'],format='%d/%m/%Y')
CASI
```

### 8. Unione delle tabelle in una temporary table ("Casi")

```
# Left-outer join su KEY_DATA poiché la table principale è quella dei casi
newCasi = pd.merge(CASI,Calendario_DAY,on='KEY_DATA',how='left')
# Per eseguire un left-outer join su 'customerid', provvediamo a ridenominare la colonna 'contactid' di CONTATTI per poi eseguire un merge
CONTATTI.rename(columns={'contactid': 'customerid'}, inplace=True)
# Eseguiamo il merge sull'attributo comune 'customerid'
newCasi = pd.merge(newCasi,CONTATTI,on='customerid',how='left')
# Per quanto riguarda la DecodificaOrigineCaso, provvediamo sempre ad eseguire prima una ridenominazione e successivamente il merge
DECODIFICA_ORIGINE_CASO.rename(columns={'ID': 'OrigineCaso'}, inplace=True)
newCasi = pd.merge(newCasi,DECODIFICA_ORIGINE_CASO,on='OrigineCaso',how='left')
# Ci troviamo davanti al caso in cui abbiamo ridondanza di informazioni (attributi duplicati). Confrontiamo KEY_DATA con DataArrivo per verificare che siano uguali, in tal caso procederemo ad effettuare un drop di una delle due
if len(newCasi[newCasi['KEY_DATA'] == newCasi['DataArrivo']]) == len(newCasi):
  print("Drop needed")
newCasi = newCasi.drop(columns=['KEY_DATA'])
# Anche il giorno della settimana è duplicato, teniamo il valore categorico piuttosto che il discreto per rappresentazioni quali hist et simil.
newCasi = newCasi.drop(columns=['DayNum'])
# Stesso discorso per OrigineCaso che può essere dedotto da Valore che rinomineremo a 'OrigineCaso'
newCasi = newCasi.drop(columns=['OrigineCaso'])
newCasi.rename(columns={'Valore': 'OrigineCaso'}, inplace=True)
# Riordiniamo le colonne in base al tipo di dato
newCasi = newCasi[['customerid', 'NumeroTicket', 'ClassificazioneTicket', 'OrigineCaso', 'day_of_week', 'DataArrivo', 'OraArrivo', 'EducationCodeValue','GenderCodeValue','RegioneDiResidenza','StatoOccupazionale']]
# Abbiamo deciso di rinominare gli attributi rendendoli 'capitalized' -->
newCasi.rename(columns={'customerid': 'CustomerID'}, inplace=True)
newCasi.rename(columns={'day_of_week': 'DayOfWeek'}, inplace=True)
# Onde evitare problemi con il riconoscimento dell'ora come stringa, trattiamo l'ora come un intero
newCasi['OraArrivo'] = pd.to_numeric(newCasi['OraArrivo'])
newCasi
```

```
# Prima riconvertiamo la struttura dati da dataframe Pandas a dataframe Spark
Casi = spark.createDataFrame(newCasi)
# Generiamo una temporary table per elaborare i dati
temp_table_name = "Casi_pandas"
Casi.createOrReplaceTempView(temp_table_name)
```

### 11a. Top 5 regioni per numero di Ticket

```
%sql
CREATE OR REPLACE TABLE topRegion
select COUNT(NumeroTicket) as `total_ticket`, `RegioneDiResidenza` as `Regione`
from `Casi_pandas`
where `OrigineCaso`="Telefono" AND `RegioneDiResidenza` IS NOT NULL
group by Regione
order by `total_ticket` DESC
LIMIT 5
```

### 11b. Percentuale di Ticket per Top 5 regioni e restanti

```
%sql
select SUM(total_ticket) as total_ticket
from (select COUNT(NumeroTicket) as `total_ticket`, `RegioneDiResidenza` as `Regione`
from `Casi_pandas`
where `OrigineCaso`="Telefono" AND `RegioneDiResidenza` IS NOT NULL
group by Regione
order by `total_ticket` ASC
LIMIT 15)
```
